# 🧠 RC Cascade Event Analysis
Comprehensive analysis of cascaded events and walking behavior in the RC task.

```

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load event summary
event_df = pd.read_csv("event_summary.csv")
```

```

# Filter to "Just dropped a pin" events
pins_df = event_df[event_df["event_type"] == "Just dropped a pin"].copy()

# Drop quality summary
print("Drop Quality Counts:")
print(pins_df["drop_quality"].value_counts())

# Score by quality
print("\nDrop Score Stats:")
print(pins_df.groupby("drop_quality")["result_score"].describe())
```

```

# Score vs Distance
plt.figure()
sns.scatterplot(data=pins_df, x="drop_distance", y="result_score", hue="drop_quality")
plt.title("Drop Score vs Distance")
plt.xlabel("Drop Distance")
plt.ylabel("Result Score")
plt.tight_layout()
plt.show()
```

```

# Bonus by drop quality
plt.figure()
sns.boxplot(data=pins_df, x="drop_quality", y="result_bonus")
plt.title("Bonus Distribution by Drop Quality")
plt.tight_layout()
plt.show()
```

```

# Walking periods
walking_df = event_df[event_df["event_type"] == "Walking Period"].copy()

# Merge with outcomes
walk_merge = walking_df.merge(
    pins_df[["cascade_id", "drop_quality", "result_score"]],
    on="cascade_id", how="left"
)

# Walking duration by outcome
plt.figure()
sns.boxplot(data=walk_merge, x="drop_quality", y="duration")
plt.title("Walking Duration by Drop Quality")
plt.tight_layout()
plt.show()

# Summary stats
print(walk_merge.groupby("drop_quality")["duration"].describe())
```

```

# --- Head Movement Analysis (if available) ---
# These assume you have 'HeadYaw', 'HeadPitch', 'HeadRoll' columns during walking

head_cols = ['HeadYaw', 'HeadPitch', 'HeadRoll']
has_pose = all(col in walking_df.columns for col in head_cols)

if has_pose:
    for col in head_cols:
        plt.figure()
        sns.histplot(data=walking_df, x=col, bins=40, kde=True)
        plt.title(f"Distribution of {col} During Walking")
        plt.tight_layout()
        plt.show()
else:
    print("Head movement data not available in walking periods.")
```
